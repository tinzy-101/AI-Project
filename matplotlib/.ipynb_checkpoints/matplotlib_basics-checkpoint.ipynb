## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.font_manager
```

## Matplotlib Basics

```python
# plots graph with points (1,2) (2,4) and (3,6)
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

## reize graph
# dpi = dots per square inch (looks better if higher dpi, but also makes size bigger)
plt.figure(figsize=(5,3), dpi=100)

# plot a straight line using x and y vectors
# label the line as "2x" in the top right corner of graph
# marker = dot marking the ordered pairs
    # plt.plot(x, y, label='2x', color='purple', linewidth=3, linestyle='--', marker='.', markersize=10)

# shorthand notation for plotting line
# r = red color
# . = marker
# - = line style (-- = dotted line)
plt.plot(x, y, 'r.-', label='2x')

## line number 2
# create array fom 0-4 w/ 0.5 steps (4.5 excluded)
x2 = np.arange(0,4.5, 0.5)
# create graph with first 5 points connected with solid line, and rest connected w/ dotted line
plt.plot(x2[:5], x2[:5]**2, 'b.-', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'b--')

# error: cannot change font when: fontdict={'fontname': 'Arial'}
plt.title("Our first graph!", fontdict={'fontsize': 20})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

# saves graph to same directory
plt.savefig('mygraph.png', dpi=300)

# get rid of [<matplotlib.lines.Line2D at 0x7f74e955de80>] line on top of graph
plt.show()

```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

# change size of bar chart
# can change title, axes titles, etc just like with line graph
plt.figure(figsize=(6,4))

# set hatch (color in bars with patterns)
bars[0].set_hatch("/")
bars[1].set_hatch("o")
bars[2].set_hatch("*")

# alternative way to hatch
'''
patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
'''


plt.show()
```
