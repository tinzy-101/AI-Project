### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8, 5))

plt.title('Gas Prices Over Time (in USD)')

# view countries' gas prices as a function of year! (x = year, y = countries)
'''
plt.plot(gas.Year, gas.USA, 'y.-', label='USA')
plt.plot(gas.Year, gas.Canada, '.-', label='Canada')
# need to use diff syntax if any field has >1 word, like SK
plt.plot(gas.Year, gas['South Korea'], 'g.-',  label='South Korea')
plt.plot(gas.Year, gas.France, 'r.-', label='France')
'''

# pandas: printing out gas years, from start to end, but skipping so only every 3rd year
print(gas.Year[::3])

# don't want to print out every gas year, b/c too squished and can't read; only want every 3rd year
plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('Price (USD)')

# iterate thru the data frame to print out all the different countries
print("\n")
for country in gas:
    print(country)

# quicker way to plot all lines on graph
# won't have all the labels unless put "label=country"
'''
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)
'''

# define own array of countries u care about, and plot only those countries using iterating method
important_countries = ['Australia', 'France', 'Mexico']

for country in gas:
    if country in important_countries:
        plt.plot(gas.Year, gas[country], marker='.', label=country)


plt.legend()


plt.savefig('Gas_price_figure.png')


plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

## Histograms`

```python
bins = np.arange(40, 110, 10)

# can use hex for color too!
plt.hist(fifa.Overall, bins=bins, color="purple")

plt.xticks(bins)

plt.ylabel("Number of Players")
plt.xlabel("Skill level")
plt.title("Distribution of Player Skills in FIFA 2018")


plt.show()
```
